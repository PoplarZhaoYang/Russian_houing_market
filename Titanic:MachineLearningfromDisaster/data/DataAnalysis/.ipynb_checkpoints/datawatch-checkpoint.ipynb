```python
import pandas as pd
from pandas import Series,DataFrame
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
%matplotlib inline

#the machine learn lib
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC, LinearSVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
```

```python
# get titanic & test csv files as a DataFrame(2D data structure in pandas)
titanic_df = pd.read_csv("../train.csv")
test_df    = pd.read_csv("../test.csv")

# preview the data
titanic_df.head()
```

```python
#show more information of DataFrame
titanic_df.info()
print("<------------------->")
test_df.info()
```

```python
titanic_df = titanic_df.drop(['PassengerId','Name','Ticket'], axis=1)
test_df    = test_df.drop(['Name','Ticket'], axis=1)
```

```python
titanic_df.describe()
```

```python
#process the missing data
titanic_df["Embarked"] = titanic_df["Embarked"].fillna("S")
#the useful plot function. size means the height and aspect is the length rate of row. kind make it like a violin.
#we can also see the S have less people who survived compared to others.
sns.factorplot(x="Embarked", y="Survived", data=titanic_df, size=4, aspect=2.8, kind="violin")
```

```python
fig, (axis1, axis2, axis3) = plt.subplots(1, 3, figsize=(15, 5))
sns.countplot(x="Embarked", data=titanic_df, ax=axis1)
sns.countplot(x="Survived", hue="Embarked", data=titanic_df, ax=axis2)

embark_per = titanic_df[["Embarked", "Survived"]].groupby("Embarked", as_index=False).mean()
#barplot
sns.barplot(y="Survived", x="Embarked", data=embark_per, order=('S', 'Q', 'C'), ax=axis3)
```

```python
#get a DataFrame that shows the Boolean matrix about Embarked
embark_dummies_titanic = pd.get_dummies(titanic_df['Embarked'])
#It seems to the S is a bad feature, so we remove it.drop the S col inplace in second dimension.
embark_dummies_titanic.drop(['S'], axis=1, inplace=True)
#the same to test
embark_dummies_test = pd.get_dummies(test_df['Embarked'])
embark_dummies_test.drop('S', axis=1, inplace=True)

#covert Embarked to two col(have deleted S)
titanic_df.drop(['Embarked'], axis=1, inplace=True)
test_df.drop(['Embarked'], axis=1, inplace=True)
#join two new col
titanic_df.join(embark_dummies_titanic)
test_df.join(embark_dummies_test)
```
